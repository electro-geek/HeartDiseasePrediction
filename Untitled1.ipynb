```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
dataset = pd.read_csv('heart.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
print(dataset.columns)
```

```python
list = dataset.columns
for i in list:
    mean_value = dataset[i].mean()
    dataset[i].fillna(value=mean_value, inplace = True)
```

```python
dataset.isnull().sum()
```

```python
#taking independent and dependent features
X = dataset.iloc[:, :-1]
y = dataset['AHD']
```

```python
#train test spliut
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
#now making a logistic regression model 
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
params = {
    "penalty":["l1","l2"]
}
lg_reg= LogisticRegression()
lg_model = GridSearchCV(lg_reg, param_grid=params, scoring = "recall", cv = 5)
```

```python
#fitting the model
lg_model.fit(X_train,y_train)
```

```python
lg_model.best_params_
```

```python
y_pred = lg_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
#making a decision tree
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
#constructing the decsion tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtc, filled='True')
```

```python
#now after seeing this we can see max_depth = 6 would be ok
treemodel = DecisionTreeClassifier(max_depth=6)
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled='True')
```

```python
tree_pred = treemodel.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(tree_pred,y_test))
print(classification_report(tree_pred,y_test))
```

```python
#this tells accuracy of logistic regression is better than decision tree
```
